````python
q = """select * from orderDetail"""

cur.execute(q).fetchall()
[c[0] for c in cur.description]
````

````python
def query_to_df(query):
    data= cur.execute(q).fetchall()
    col_names = [c[0] for c in cur.description]
    return pd.DataFrame(data,columns=col_names)
````

````python
q = """select * from orderDetail"""
orderDetail = query_to_df(q)
orderDetail
````

````python
orderDetail['discounted'] = orderDetail["Discount"]>0
orderDetail['discounted'] = orderDetail['discounted'].map({True:'Discounted',
                                                           False:'Full Price'})
orderDetail
````

````python
sns.barplot(data=orderDetail, x='discounted',y='Quantity')
````
